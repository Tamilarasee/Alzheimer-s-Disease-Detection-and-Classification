1.  Explain the differences between supervised, semi-surpervised, weakly-supervised and unsupervised learning.



**Answer:**


Supervised learning is a type of machine learning in which a machine is trained using a training set of input and output data. Based on the training data, the machine identifies the relation between input and output as they are known (labeled) and this relation/function is later on used to predict the output when an unseen input sample is given. Regression and Classification are types of Supervised learning.

Unsupervised learning is one where the machine is fed with only input data and no corresponding output data for training.The goal of this learning method is to recognize similar patterns in the given inputs and classify them as different groups (the group name is also known).It is designed to analyze and extract useful information from the hidden structures in the provided unlabeled data. Its applications are Clustering, density estimation, visualization,assocaiation

Semi-supervised learning is a combination of Supervised and Unsupervised learning introduced to reduce the cost of labeling all the data in the training set as in case of Supervised and the ambiguity in Unsupervised learning. In Semi-supervised learning, the machine is trained with a combination of a few labeled data and more of unlabeled data. It aims to arrive at a function/relation between the input and output variables.It is commonly used for Text/Image classification.

Weakly-supervised learning lies in between Supervised and Semi-supervised learning. It is used to train a machine with large amount of weakly labeled data generated from noisy, inaccurate sources.The weak labels are usually generated by computer by using strategies to combine new unlabeled data with a signal to create their label.This helps reduce data shortage.Object recognition is one of its applications.


The mentioned machine learning types have their main difference in what type of training set is provided and how it is handled by the machine to arrive at useful information from the data.

3. Explain how to perform N-fold cross-validation.



**Answer:**
Cross-Validation is a technique for evaluating the performance of a given data model. We can run the technique on various models and pick the most appropriate model based on the performance comparison

Steps:

1. Reaarange the given dataset randomly
2. Split the dataset into N equal folds(groups)
3. Pick one fold as Test Set and consider the remaining N-1 folds as Training Set
4. Pick a model to evaluate and fit it on the Training Set, validate on the Test Set and save the validation score and discard the model
5. Repeat the steps 3 & 4 N times for the same model, but pick a different fold as Test set each time such that every sample had the chance to be on the Test set once and on the training set for N-1 times.
6. The average of N validation results is the final score for a given model.
7. Repeat the same for different models and compare the average scores of every model to identify the best model for the machine.

**Answer:**

The predicted output y for all i =  1 to m samples,

where i refers to the ith sample in the training set and n is number of features

To simplify our notation, we can also introduce the convention of letting $x_{0} = 1$ (this is the intercept term)

$$h_{\theta}(x^{(i)}) = \hat y^{(i)}=\sum_{j=0}^{n} \theta_{j}x_{j}^{(i)} \quad for \space all \space i = 1 \space to \space m \space samples$$



7. Write the objective function that can be used to determine the regression model parameters. How is this objective function will be used to find model parameters?



**Answer:**

The loss function (cost function) is th eobjective function that defines the expression corresponding to the difference in the predicted output and the original output. This loss function must be a minimum value which would imply that there is less variation from the required output.

Loss function, $$L(\theta) = \frac{1}{2m} \sum_{i=1}^{m} (h_{\theta}(x^{(i)}) - y^{(i)})^2 $$

To minimize the loss function, we have to determine the appropriate values of the parametes $\theta_{j} \space for \space j = \space 0 \space to \space n$

To find the parameters,we use the least mean square (LMS) rule as given by below expression,
$$\theta_{j} = \theta_{j} - \alpha \frac {\partial}{\partial \theta_{j}} L(\theta)$$
where $\space \alpha $ is the learning rate and the partial derivative term correspond to the gradient descent algorithm.

This can be further given by below expression for all i=1 to m samples and j=0 to n parameters,
$$\theta_{j} = \theta_{j} - \alpha (h_{\theta}(x^{(i)}) - y^{(i)})(x_{j}^{(i)})$$

```python
# initialize theta
# we need random values for theta_0 and theta_1
# we use randin to generate a value with zero mean so that the initialization is not biased in any direction
slr_theta = np.random.randn(2,1)
print(f'Theta values: {slr_theta[0]}, {slr_theta[1]}')
```

```python
# As per formula, hypothesis, h(x) = theta_0 +theta_1 (x)
def slr_hypothesis(theta,X):
    h = theta[0] + (theta[1]*X)
    return h
```

```python
# define cost function
def slr_cost_function(X,y,theta):
    h = slr_hypothesis(theta,X)
    m = y.size
    J = 1/(2*m)*np.sum((h-y)**2)
    return J
```

```python
# calculate cost by calling the function and print cost results
slr_cost = slr_cost_function(slr_train_X,slr_train_y,slr_theta)
print(f'Given: theta_0 = {slr_theta[0]}, theta_1 = {slr_theta[1]} \nCost = {slr_cost}')
```

**Question: How did you choose your lr and epoch number?**

**Answer:**

Tried multiple values of learning rate(alpha) and epochs and identified the best values on comparison for training and testing cost. Based on the comparison,the learning rate is set to be 1 in order to make the optimisation procedure more stable and effective, and epoch number to be 1500 to make sure the model will converge therby having reasonable trraining and testing cost.

```python
# gradient descent to find the optimal fit
m = slr_train_y.size
def slr_gradient_descent(epoch,alpha,slr_train_X,slr_train_y,theta):
    iterator = 0
    cost_history = []
   
    while iterator < epoch:
        h = slr_hypothesis(theta,slr_train_X)
        theta[0] -= (alpha/m)*np.sum(h-slr_train_y)
        theta[1] -= (alpha/m)*np.sum((h-slr_train_y)*slr_train_X)
        cost_history.append(slr_cost_function(slr_train_X,slr_train_y,theta))
        iterator += 1
    return theta,cost_history
epoch = 1500
alpha = 1
slr_theta = np.zeros(2)
slr_theta,slr_cost_history = slr_gradient_descent(epoch,alpha,slr_train_X,slr_train_y,slr_theta)
print(f'Given:\nepoch--> {epoch}\nalpha--> {alpha} ')
print(f'\nFinal values:\nTheta--> {slr_theta[0]}, {slr_theta[1]}\nCost--> {slr_cost_history[-1]}')


# Used below code to find the optimal alpha and epoch values

# epoch_list = [100,500,1000,1500,2000,2500,3000]
# alpha_list = [1.5,1,0.8,0.5,0.3,0.1,0.03,0.01,0.003,0.001]

# combination={}
# for epoch in epoch_list:
#     combination[epoch] = []
#     for alpha in alpha_list:
#         slr_theta = np.zeros(2)
#         slr_theta,slr_cost_history = slr_gradient_descent(epoch,alpha,slr_train_X,slr_train_y,slr_theta)
#         mse_test = slr_cost_function(slr_test_X,slr_test_y,slr_theta)
#         combination[epoch].append(f'{alpha} - {slr_cost_history[-1]} - {mse_test}')
# print(combination)
```
